```python
import pandas as pd 
from chembl_webresource_client.new_client import new_client
```

### Search For Target Protein

##### Coronavirus Target Search

```python
target = new_client.target
target_query = target.search("coronavirus")
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
#selecting and retrieveing bioactivity data for "SARS coronavirus"

selected_target = targets.target_chembl_id[6]
print("Unique ID for Target: ", selected_target)
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
#standard type ensures that we do not get a mixture of activity measurements
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.head()
```

```python
df.to_csv('../../data/bioactivity_data.csv', index=False)
```

##### Handling missing data

if any comounds has missing values for the standard_value col, we drop 

```python
df_filtered = df[df.standard_value.notna()]
df_filtered
```

##### Data pre-processing of the bioactivity data

```python
bioactivity_class = []

for i in df_filtered.standard_value:
    if float(i) >= 10_000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1_000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")
```

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
```

```python
pd.Series(bioactivity_class).unique()
```

```python
df_3 = df_filtered[selection].reset_index(drop=True)  # reset index for alignment
bioactivity_class_series = pd.Series(bioactivity_class, name="bioactivity_class")

df_final = pd.concat([df_3, bioactivity_class_series], axis=1)
df_final
```

```python
df_final.to_csv("../../data/bioactivity_processed_data.csv", index=False)
```
